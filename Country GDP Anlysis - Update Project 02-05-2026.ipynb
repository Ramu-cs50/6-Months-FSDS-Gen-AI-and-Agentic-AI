***We analyse  pythoon datafram or dataset with seaborn***

```python
#Seaborn is a powerful statistical visualisation pkg in python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']=6,2
#seaborn are used for advannced graphs like line,distribution plot etc
import warnings
warnings.filterwarnings('ignore')
df.columns
```

```python
df['InternetUsers']
```

```python
vis1=sns.distplot(df['InternetUsers'])
#univariate analysis -   Plotting the graph using 1 variable is known a univariate analysis
```

```python
vis2=sns.displot(df['InternetUsers'])
#univariate analysis -   Plotting the graph using 1 variable is known a univariate analysis
```

```python
vis3=sns.distplot(df['InternetUsers'], bins=15)
```

```python
#box plot in stats and booxplot is there only in sns
#It is a bivariate analysis because it is using 2 variables
vis4=sns.boxplot(data = df, x='IncomeGroup', y ='BirthRate')
```

```python
#Outliers are the datapoints which stand away from the 
vis5=sns.lmplot(data=df,x='InternetUsers', y='BirthRate')
```

```python
#tO REMOVE THE LINE IN THE SCATTER PLOT WE USE FIT_REG=fALSE
vis5=sns.lmplot(data=df,x='InternetUsers', y='BirthRate', fit_reg=False)
```

```python
#hue is like a labes in the dataset for the data set
vis5=sns.lmplot(data=df,x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup')
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers', y='BirthRate', fit_reg=True, hue='IncomeGroup')
```
