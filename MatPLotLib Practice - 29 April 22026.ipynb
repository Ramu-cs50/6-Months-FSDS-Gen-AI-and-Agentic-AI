```python
#MAtplotlib practice
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1=np.linspace(0,10,50)
```

```python
x1
```

```python
plt.plot(x1, np.sin(x1),ls='--', color='green', marker='o', label="Sinx1")
plt.plot(x1, np.cos(x1), '-.', color='blue', marker='s', label="Cosx1")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.subplot(3,1,1) # (rows, columns, panel number)
plt.plot(x1,np.cos(x1), '*')
```

```python
#Here trying to plot 3 graphs ijn a single plot for comparision purposes
#create a plot figure
plt.subplot(3, 1, 1)
plt.plot(x1, np.sin(x1), '--')
plt.title("Sinx")
plt.subplot(3, 1, 2)
plt.plot(x1, np.cos(x1), '--')
plt.title("Cosx")
plt.subplot(3, 1, 3)
plt.plot(x1, np.tan(x1), '--')
plt.title("Tanx")
```

```python
x=np.linspace(1, 5, 20)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.plot()
plt.legend()
```
